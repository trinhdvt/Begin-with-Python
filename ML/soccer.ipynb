```python
# Custom select
df1 = pd.DataFrame(data_frame,columns=['Name','Wage','Value'])
df1.head() # First five row

def value_to_float(x:str):
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1e3
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1e6
    if 'B' in x:
        if len(x) > 1:
            return float(x.replace('B', '')) * 1e9
    return 0.0

wage = df1['Wage'].replace(r'[€]','',regex=True).apply(value_to_float)
value = df1['Value'].replace(r'[€]','',regex=True).apply(value_to_float)

df1['Wage'] = wage
df1['Value'] = value

df1['diff'] =  df1['Value'] - df1['Wage']
df1.sort_values(['diff'] ,ascending=[0])
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='Wage', y='Value', data=df1)
graph
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool

TOOLTIPS = [
        ("index", "$index"),
        ("(Wage, Value)", "(@Wage, @Value)"),
        ("Name", "@Name"),
]

hover = HoverTool(tooltips=TOOLTIPS)

p = figure(title='Soccer 2019', x_axis_label='Wage', y_axis_label='Value', plot_width=600, plot_height=600,tools=[hover])
p.circle('Wage', 'Value', size=10, source=df1)
show(p)
```

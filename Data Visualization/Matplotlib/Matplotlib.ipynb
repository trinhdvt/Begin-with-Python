# Basic Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,20)
y = x ** 2
```

```python
plt.plot(x, y, "r-")
plt.xlabel("X label")
plt.ylabel("Y lable")
plt.title("Title")
plt.show()
```

### Plot multiple graphs

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "r-")

plt.subplot(1, 2, 2)
plt.plot(y, x, "b-")
```

### Use figure

> figure look like a canvas, it's contain many Axes. Each axes can use to plot.

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig ,axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
plt.tight_layout()
""" for cur_ax in axes:
    cur_ax.plot(x, y) """

axes[0].plot(x, y)
axes[0].set_title("First Plot")

axes[1].plot(y, x)
axes[1].set_title("Second Plot")
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x, x**2, label="X^2")
axes.plot(x, x**3, label="X^3")
axes.legend(loc=0)
```

## Plot appearance

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x, y, color="red", lw=2, ls="--", marker="*", markersize = 10,
            markerfacecolor="blue")

axes.set_xlim([0, 8])
axes.set_ylim([0, 20])
```

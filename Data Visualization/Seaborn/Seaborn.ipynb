# Basic Seaborn

## Distribution Plot

```python
import seaborn as sns; sns.set()
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

### Histogram plot

```python
sns.set_style("darkgrid")
sns.distplot(tips['total_bill'],kde=False, hist_kws={
    "color":"r"
    })
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
# sns.jointplot(x='total_bill', y='tip', data=tips, kind="hex")
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind="reg")
# sns.jointplot(x='total_bill', y='tip', data=tips, kind="kde")
```

```python
sns.pairplot(data=tips, hue="sex", palette="coolwarm")
```

```python
sns.rugplot(tips['total_bill'])
```

## Categorical Plot

```python
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', hue='size', data=tips)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

## Maxtrix Plot

```python
flight = sns.load_dataset("flights")
flight.head()
```

```python
fp = flight.pivot_table(values='passengers', index='month', columns='year')
```

```python
fp
```

```python
sns.heatmap(data=fp, cmap='coolwarm', linecolor='black', linewidths=1)
```

## Regression plot

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='seismic')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```
